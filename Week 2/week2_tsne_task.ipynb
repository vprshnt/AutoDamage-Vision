## Objective
Visualize high-dimensional image data in 2D using **t-SNE**.

**Note:** t-SNE is for visualization only.


```python
# TODO: Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.datasets import load_digits
```

## Step 1: Load Dataset

```python
# TODO: Load digits dataset again
from tensorflow.keras.datasets import fashion_mnist

(X_train, y_train), (X, Y) = fashion_mnist.load_data()
```

## Step 2: Preprocess

```python
# Flatten images
X_flat = X.reshape(len(X), -1)

# Normalize pixel values
X_flat = X_flat / 255.0
```

## Step 3: Apply t-SNE

```python
# TODO: Initialize TSNE with n_components=2
# Fit and transform data

tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X_flat)
```

## Step 4: Visualization

```python
plt.figure(figsize=(8, 6))
scatter = plt.scatter(
    X_tsne[:, 0],
    X_tsne[:, 1],
    c=Y,
    cmap='tab10',
    s=15,
    alpha=0.7
)

plt.colorbar(scatter, label='Class label')
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.title('t-SNE Visualization of Image Embeddings')
plt.show()
```

## Reflection
- Do clusters correspond to labels?
- Why should t-SNE not be used for training models?

t-SNE is excellent for visualizing class separability but unsuitable as a feature space for predictive modeling.
