```python
# TODO: Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_digits
```

## Step 1: Load Dataset

```python
# TODO: Load digits dataset
# X should contain images
# y should contain labels
from tensorflow.keras.datasets import fashion_mnist

(X_train, y_train), (X, Y) = fashion_mnist.load_data()
```

## Step 2: Flatten Images

```python
# TODO: Flatten images into (num_samples, num_features)
X=X.reshape(len(X), -1)
```

## Step 3: Apply PCA

```python
# TODO: Initialize PCA with n_components = 50
# Fit PCA on flattened images
pca = PCA(n_components=50)
pca.fit(X)
```

## Step 4: Reconstruction

```python
# TODO: Transform and inverse_transform using PCA
# Transform to PCA space
X_pca = pca.transform(X)

# Reconstruct back to original feature space
X_reconstructed = pca.inverse_transform(X_pca)
```

## Step 5: Visualization

```python

n_images = 5  # number of images to display
img_h, img_w = 28, 28  # change if needed

plt.figure(figsize=(2 * n_images, 4))

for i in range(n_images):
    # Original image
    plt.subplot(2, n_images, i + 1)
    plt.imshow(X[i].reshape(img_h, img_w), cmap='gray')
    plt.axis('off')
    if i == 0:
        plt.title("Original")

    # Reconstructed image
    plt.subplot(2, n_images, i + 1 + n_images)
    plt.imshow(X_reconstructed[i].reshape(img_h, img_w), cmap='gray')
    plt.axis('off')
    if i == 0:
        plt.title("Reconstructed")

plt.tight_layout()
plt.show()
```

## Reflection
- What information is lost during PCA?
- Why does reconstruction blur fine details?

PCA keeps the most important global patterns but sacrifices local fine details to achieve dimensionality reduction.
